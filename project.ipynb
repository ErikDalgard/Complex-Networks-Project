```python
import networkx as nx
import info_functions as f
import generate_networks as g
import importlib
```

### Reading the different networks

```python
#Reading the data network
G = nx.read_edgelist("data/ca-GrQc.txt", create_using=nx.Graph(), nodetype=int)

G.graph['Name'] = "GrQc"

#Creating a random network with the same number of nodes
n_nodes = len(G.nodes())
n_edges = len(G.edges())

p = n_edges/(n_nodes*(n_nodes-1))
GILBERT = g.Gilbert(n_nodes, p)

#Creating a configuration model with the same degree sequnce
degree_sequence = [d for _, d in G.degree()]
CONFIG = g.configuration_model(degree_sequence)

WATZ = g.watts_strogatz(n_nodes, 6, 0.1)

#Creating barabasi albert network
m = max(1, round(n_edges / n_nodes))
BA = g.barabasi_albert(n_nodes, m)
```
